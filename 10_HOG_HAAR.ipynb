HAAR (AdaBoosted Classification Haar-like Features) es un método de detección de características utilizado en visión por computadora para detectar objetos en imágenes. El método HAAR utiliza una serie de características de "Haar-like" para detectar patrones y formas específicas en una imagen. Estas características son pequeñas regiones de una imagen que se suman o restan entre sí para formar una medida de la presencia o ausencia de un patrón específico. Luego, se utiliza un clasificador para determinar si el patrón está presente en la imagen.

```python
import cv2
import matplotlib.pyplot as plt

# Carga la imagen
imagen = cv2.imread('Gente.jpg')
#imagen = cv2.resize(imagen, (0,0), fx=0.5, fy=0.5)

# Carga el clasificador HAAR para detectar rostros
clasificador = cv2.CascadeClassifier('haarcascade_frontalface_alt2.xml')

# Detecta rostros en la imagen
rostros = clasificador.detectMultiScale(imagen)

# Dibuja un rectángulo alrededor de cada rostro detectado
for (x, y, ancho, alto) in rostros:
    cv2.rectangle(imagen, (x, y), (x + ancho, y + alto), (0, 255, 0), 2)

# Muestra la imagen con los rostros detectados con plt 
plt.imshow(imagen[:,:,::-1])
plt.show()
```

HOG (Histogram of Oriented Gradients) es otro método de detección de características utilizado en visión por computadora para detectar objetos en imágenes. El método HOG utiliza un histograma de gradientes orientados para describir la forma y la orientación de las características en una imagen. Los gradientes orientados son medidas de cambio en el brillo de una imagen en una dirección específica. Al calcular el histograma de estos gradientes orientados en una imagen, se puede obtener una descripción robusta de las características de la imagen que se puede utilizar para detectar objetos.

```python
import cv2

# Carga la imagen
imagen = cv2.imread('Gente.jpg')

# Inicializa el descriptor HOG
descriptor_hog = cv2.HOGDescriptor()

# Carga el archivo de pesos del clasificador HOG
descriptor_hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())

# Detecta personas en la imagen
(rects, weights) = descriptor_hog.detectMultiScale(imagen, winStride=(4, 4), padding=(8, 8), scale=1.05)

# Dibuja un rectángulo alrededor de cada persona detectada
for (x, y, ancho, alto) in rects:
    cv2.rectangle(imagen, (x, y), (x + ancho, y + alto), (0, 255, 0), 2)

# Muestra la imagen con las personas detectadas con plt
plt.imshow(imagen[:,:,::-1])
plt.show()
```
